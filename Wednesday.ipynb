```python
def weather_api(year):    
    daily_high_temp = {}
    url = 'https://api.darksky.net/forecast/'
    ny_coord = '/40.7829,-73.9654,'
    midday = 'T12:00:00'
    
    for i in range(1,13):
        print(i)
        days = monthrange(year, i)[1]
        
        for j in range(1, days+1):
            if i < 10:
                month = '0'+ str(i)
            
            else:
                month = str(i)
            
            if j < 10:
                day = '0' + str(j)
            
            else:
                day = str(j)
            
            r = requests.get(url + darksky_key2
                             + ny_coord + str(year)
                             + '-' + month + '-' + day + midday)
            key = str(year) + '-' + month + '-' + day
            value = r.json()['daily']['data'][0]['temperatureHigh']
            
            daily_high_temp[key] = value
    
    return pd.DataFrame.from_dict(daily_high_temp, orient='index')
```

```python
df_temp_2014 = weather_api(2014)
```

```python
df_temp_2006 = weather_api(2006)
df_temp_2006.to_csv('df_temp_2006.csv')
```
